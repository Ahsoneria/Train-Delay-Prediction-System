```
def merge_jsons(list_of_jsons):
  jsons = []
  for js in list_of_jsons:
    with open(js,'r') as file:
      jsons.append(json.loads(file.read()))
  final = {}
  for js in jsons:
    final = {**final, **js}
  return final
```
